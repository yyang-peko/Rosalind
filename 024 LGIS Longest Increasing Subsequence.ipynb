```python
def findLIS(A):
 
    # `LIS[i]` stores the longest increasing subsequence of sublist
    # `A[0i]` that ends with `A[i]`
    LIS = [[] for _ in range(len(A))]
 
    # `LIS[0]` denotes the longest increasing subsequence ending at `A[0]`
    LIS[0].append(A[0])
 
    # start from the second element in the list
    for i in range(1, len(A)):
 
        # do for each element in sublist `A[0i-1]`
        for j in range(i):
 
            # find the longest increasing subsequence that ends with `A[j]`
            # where `A[j]` is less than the current element `A[i]`
 
            if A[j] < A[i] and len(LIS[j]) > len(LIS[i]):
                LIS[i] = LIS[j].copy()
 
        # include `A[i]` in `LIS[i]`
        LIS[i].append(A[i])
 
    # `j` will store the index of LIS
    j = 0
    for i in range(len(A)):
        if len(LIS[j]) < len(LIS[i]):
            j = i
 
    # print LIS
    print(*LIS[j], sep= ' ')
```
