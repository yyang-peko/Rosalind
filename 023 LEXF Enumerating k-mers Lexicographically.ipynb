```python
a = 'A B C D'
n = 4
```

```python
chars = a.split()
```

```python
results  = []
for j in itertools.product(chars, repeat = n):
    string = ''.join(j)
    results.append(string)
results.sort()
```

```python
for i in results:
    print(i)
```
