```python
a = open('010 rosalind_cons.txt', 'r').read()
```

```python
b = a.replace('\n', '').split('>')[1:]
sequences=[]
for i in b:
    seq = i[13:]
    sequences.append(seq)
```

```python
import pandas as pd
```

```python
s = []
for i in sequences:
    s.append(list(i))
    
df = pd.DataFrame(s)
```

```python
df
```

```python
counts = pd.DataFrame(columns = ['A', 'C', 'G', 'T'])
consensus = ''

for i in df:
    counts=counts.append(df[i].value_counts())
    consensus += (df[i].value_counts().idxmax())

counts=counts.fillna(0).astype(int)
```

```python
print(consensus)
print(counts.T)
```

```python
with open('010 output.txt', 'a') as f:
    dfAsString = counts.T.to_string(header=False, index=True)
    f.write(dfAsString)
```
