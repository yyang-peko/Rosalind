```python
def overlap(a, b, min_length = 4):
    start = 0
    while True:
        start = a.find(b[0:min_length], start)
        if start == -1:
            return 0
        if b.startswith(a[start:]):
            return len(a) - start
        start +=1
```

```python
def max_overlap(reads, k):
    reada, readb = None, None
    best_olen = 0
    for a, b in itertools.permutations(reads, 2):
        olen = overlap(a, b, min_length = k)
        if olen > best_olen:
            reada, readb = a, b
            best_olen = olen
    return reada, readb, best_olen
```

```python
def greedy_scs(reads, k):
    read_a, read_b, olen = max_overlap(reads, k)
    while olen > 0:
        reads.remove(read_a)
        reads.remove(read_b)
        reads.append(read_a + read_b[olen:])
        read_a, read_b, olen = max_overlap(reads, k)
    return ''.join(reads)
```
